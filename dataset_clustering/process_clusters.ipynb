```python
import re
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv("/dls/science/groups/i04-1/conor_dev/dataset_clustering/output/clusters_kind_of_working.csv")
```

```python
df
```

```python
df["system"] = df.apply(lambda row: re.match("^[^-]+", str(row.dtag)).group(), 
                        axis=1,
                       )
```

```python
df
```

```python
clusters_by_system = []
for system in pd.unique(df["system"]):
    print(system)
    
    clusters = pd.unique(df[df["system"] == system]["cluster"])
    
    print(df[df["system"] == system])
    
    if len(df[df["system"] == system]) <2:
        continue
    
    try:
        clusters_by_system.append({"system": system,
                                  "num_clusters": max(clusters[clusters > -1]),
                                  }
                                 ) 
    except:
        print("\tWARNING: system has no clusters: {}".format(system))
#     print(clusters_by_system)
#     print(clusters)
```

```python
clusters_by_system
```

```python
system_clusters_df = pd.DataFrame(clusters_by_system)
```

```python
system_clusters_df
```

```python
plt.figure(figsize=(15, 15))
sns.barplot(x="system",
           y="num_clusters",
           data=system_clusters_df,
           )
```

```python
plt.figure(figsize=(15, 15))

sns.distplot(system_clusters_df["num_clusters"])
```
